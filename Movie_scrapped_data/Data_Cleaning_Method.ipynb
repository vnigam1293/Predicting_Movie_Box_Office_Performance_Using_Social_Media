```python
def dataCleaning(df):
    df.drop(['date'],axis=1)
    dataList = df['tweet'].values
    string = str(tuple([str(e) for e in dataList]))
    tokens_tweet = [t for t in nltk.sent_tokenize(string)]
    word_tokens = [t_tokenizer.tokenize(i) for i in tokens_tweet]
    
    t = []

    for w in word_tokens:
        word = []
        for wr in w:
            if wr not in newstopwords and wr.isalpha():
                word.append(wr.lower())
        t.append(word)
    
    sentences = []
    for l in t:
        sentences.append(TreebankWordDetokenizer().detokenize(l))
    
    return sentences 
```

```python
word_tokens_list = dataCleaning(m47)
```

```python
word_tokens_list
```

```python
def sentimentAnalysis(df,sentences):
    pos_comments = [comment  + "" +',pos' for comment in sentences if sia.polarity_scores(comment)['compound'] > 0 ]
    neg_comments = [comment  + "" +',neg' for comment in sentences if sia.polarity_scores(comment)['compound'] < 0 ]
    percent_pos_comments = len(pos_comments)/len(pos_comments + neg_comments)

    percent_neg_comments = len(neg_comments)/len(pos_comments + neg_comments)
    df['pos_sentiment'] = percent_pos_comments
    df['neg_sentiment'] = percent_neg_comments
    
    return percent_pos_comments, percent_neg_comments
```
